# Time Related and Customer Profiling

# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading File

```python
df = pd.read_csv(r"E:\E commerce project\Cleaned_Ecommerce_Data _working.csv")
```

# Checking when do Customers purchase the most?

```python
# Comparing Weekday vs Weekend purchases (from Is_Weekend or Purchase_Day) 
```

```python
# Made sure Is_Weekend is boolean
df['Is_Weekend'] = df['Is_Weekend'].astype(bool)

# Plotted purchase count
df['Is_Weekend'].value_counts().rename({True: 'Weekend', False: 'Weekday'}).plot(kind='bar', color='coral')
plt.title('Purchase Count: Weekend vs Weekday')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# customers purchase the most on weekdays. 
```

```python
# Checking whether Social Media Influence changes across different demographics
```

```python
# Analyzing how it varies across different demographics like:
#    - Gender
#    - Age
#    - Income_Level
#    - Education_Level
#   - Occupation
```

```python
# Social_Media_Influence on gender

sns.countplot(data=df, x='Gender', hue='Social_Media_Influence')
plt.title("Social Media Influence by Gender")
plt.show()
```

```python
# Both males and females show high counts across all influence levels.
# "High" influence is slightly higher for males than females.
# Other gender identities have lower sample sizes, but influence levels are more evenly spread.
```

```python
# Social_Media_Influence on age (using boxplot)

sns.boxplot(data=df, x='Social_Media_Influence', y='Age')
plt.title("Age Distribution by Social Media Influence")
plt.show()
```

```python
# All influence categories (unknown, medium, low, high) have very similar age distributions.
# Median ages stay around the mid-30s across all influence levels.
```

```python
# Social_Media_Influence on Income_Level

sns.countplot(data=df, x='Income_Level', hue='Social_Media_Influence')
plt.title("Social Media Influence by Income level")
plt.show()
```

```python
# Both middle and high-income groups show.
# "High" social media influence has the most users
# Followed by "unknown", "low", and "medium"
# The pattern is similar across both income groups.
```

```python
# Social_Media_Influence on Education_Level

sns.countplot(data=df, x='Education_Level', hue='Social_Media_Influence')
plt.title("Social Media Influence by Education Level")
plt.xticks(rotation=15)
plt.show()
```

```python
#  High influence is the tallest bar across all education levels.
#  Influence is fairly even across levels — from school to postgraduate.
#  There's no drastic difference in influence patterns between education levels.
```

```python
# Social_Media_Influence on Occupation

sns.countplot(data=df, x='Occupation', hue='Social_Media_Influence')
plt.title("Social Media Influence by Occupation")
plt.show()
```

```python
#  High influence is common across most occupations (students, professionals, unemployed, etc.).
#  Students may show slightly more high influence than other groups (if bar is taller).
#  Other occupation groups (e.g., professionals, unemployed) have balanced but lower counts.
```
